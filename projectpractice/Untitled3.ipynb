```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({
     "Gender": ["Male", "Female", "Male", None],
    "Education": ["Graduate", "Not Graduate", "Graduate", "Graduate"],
    "LoanAmount": [100, 150, np.nan, 200],
    "Loan_Status": ["Y", "N", "Y", "N"]})
df
   
```

```python
#sum of null values based on the columns
df.isnull().sum()
```

```python
#fillig missing values 
df.ffill(inplace=True)
df.bfill(inplace=True)
df 

```

```python
#ML models cannot read text like "Male","Graduate"
df1=pd.get_dummies(
    df,columns=[
        "Gender","Education","Loan_Status"])
df1=df1.astype(int)
df1
```

```python
#to drop unnecessary tables
df1.drop(["Gender_Male", "Education_Not Graduate", "Loan_Status_Y"],
         axis=1,
         inplace=True,
         errors="ignore")
print(df1.columns)
```

```python
df = pd.read_csv(r"C:\Users\jacob\Downloads\loan_data_set_lyst1728376986423 (1).csv")
df.ffill(inplace=True)
df.bfill(inplace=True)
df
```

```python
df1=pd.get_dummies(df,columns=[
     "Gender", "Married", "Dependents",
        "Education", "Self_Employed",
        "Property_Area", "Loan_Status"
    ]
                  )
df1
```

```python
df1.drop("Gender_Female", axis=1, inplace=True)
df1.drop("Education_Graduate", axis=1, inplace=True)
df1.drop("Loan_Status_N", axis=1, inplace=True)
df1
```

```python
#iloc=index based selection
#: selectall
x=df1.iloc[:, :-1]#Data used to learn (features)
y=df1.iloc[:,-1]#Data used to learn (features)
print(x)
print(y)
```

```python
#X  → Inputs / questions

#Y  → Output / answer you want the computer to learn
from sklearn.model_selection import train_test_split
#x_train, y_train	Teach the model
#x_test, y_test	Test the model
#test_size=0.25
#➡ 25% data for testing
#➡ 75% data for training
#random_state=0
#➡ Ensures same split every time
#x_train	Questions the model learns from
#y_train	Correct answers for learning
#x_test	 New questions (exam)
#y_test	  Correct answers (answer key)
#y_test
# Actual / real answers
#(from the dataset)

#y_pred
#Predicted answers
#(from the ML model)

x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.25,random_state=0
)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=(0))

from sklearn.linear_model import LogisticRegression
m1=LogisticRegression()
m1.fit(x_train, y_train)
y_pred=m1.predict(x_test)

from sklearn.metrics import accuracy_score,confusion_matrix
acc=accuracy_score(y_test, y_pred)
c=confusion_matrix(y_test,y_pred)
print(acc)
print(c)
```
