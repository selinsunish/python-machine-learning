```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "continent": ["Asia", "Europe", "Asia", "Africa", np.nan],
    "location": ["India", "France", "China", "Nigeria", "Brazil"],
    "date": ["2020-01-01", "2020-01-02", "2020-01-03", "2020-01-04", "2020-01-05"],
    "total_cases": [100, 200, 150, 80, 50],
    "total_deaths": [1, 5, 3, 2, 1],
    "gdp_per_capita": [2000, 30000, 10000, 1500, 8000],
    "human_development_index": [0.65, 0.89, 0.76, 0.53, 0.71]
})
#difference
print(df)
df
print(df.shape)
print(df.dtypes)
print(df.info())
print(df.describe())
```

```python
#count unique values in columns
df['location'].value_counts()
```

```python
#max frequency
df['continent'].value_counts().idxmax()
```

```python
#max and mean of total cases
print(df['total_cases'].max())
print(df['total_cases'].mean())
```

```python
#continent with max HDI
#df.loc[row_condition, column_name]

df.loc[df['human_development_index'].idxmax(),'continent']
```

```python
#data cleaning
print(df)
df.drop_duplicates(inplace=True)
print(df)
```

```python

print(df.isna().sum())
```

```python
#drop rows where continent is missing
print(df.dropna(subset=['continent'],inplace=True))
df
```

```python
#fill remaining NaN with 0
df.fillna(0,inplace=True)
df
```

```python
print(df.dtypes)
df['date']=pd.to_datetime(df['date'])
print(df.dtypes)
```

```python
#to group the data with continent max value
df_groupby=df.groupby('continent').max().reset_index()
df_groupby
```

```python

print(df)
df_groupby['total_deaths_to_total_cases'] = (
    df_groupby['total_deaths'] / df_groupby['total_cases']
)
df
```
