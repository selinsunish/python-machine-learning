```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

```python
#scalling done here for improving the accuracy
x_train=x_train/255
x_test=x_test/255
```

```python
x_train[0]
```

```python
#Neutral networks cannot undrstand images directly
xtrain_flattened=x_train.reshape(len(x_train),28*28)
x_test_flattened=x_test.reshape(len(x_test),28*28)

x_test_flattened.shape
xtrain_flattened.shape

```

```python
xtrain_flattened[0]
```

```python
#defining the neutral networks
Sequential → layers stacked one after another

#Dense(10) →
#10 neurons = one neuron per digit (0–9)
#input_shape=(784,) →
#Each image is a vector of 784 pixels
#activation='sigmoid' →
#Each neuron outputs a value between 0 and 1

model=keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')
])
#optimizer='adam'	How the model updates weights
#loss	Measures how wrong the model is
#sparse_categorical_crossentropy	Correct loss for labels like 0,1,2,...9
#accuracy	% of correct predictions
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
#training actually happens 
#epochs=how many times
model.fit(xtrain_flattened,y_train,epochs=5)
```

```python
model.predict(x_test_flattened)
```

```python
plt.matshow(x_test[1])
```

```python
y_predicted=model.predict(x_test_flattened)
y_predicted[1]
```

```python
#picks the index of the largest value
np.argmax(y_predicted[1])
```

```python
y_predicted_labels=[np.argmax(i) for i in y_predicted]
y_predicted_labels[:5]
```

```python
#Rows → Actual digit
#Columns → Predicted digit
#Diagonal values → correct predictions
#Off-diagonal → mistakes
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm                         
                         
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

 LOST AND COST FUNCTIONS
 This followig code gives the working of errors in deep learning manually.It ca be automatically done through tensorflow and keras

```python
y_predicted=np.array([1,1,0,0,0,1])
y_true=np.array([0.30,0.7,1,0,0.5])

                     
```

```python
def mae(y_true,y_predicted):
    total_error=0
    for yt,yp in zip(y_true,y_predicted):
        total_error+=abs(yt-yp)
    print("Total error:",total_error) 
    mae=total_error/len(y_true)
    print("mae",mae)
    return mae
       
```

```python
mae(y_true,y_predicted)
```

```python
epsilon=1e-15
```

```python
y_predicted_new=[max(i,epsilon)for i in y_predicted]
y_predicted_new
```

```python
def log_loss(y_true,y_predicted):
    epsilon=1e-15
    y_predicted_new=[max(i,epsilon) for i in y_predicted]
    y_predicted_new=[min(i,1-epsilon) for i in y_predicted_new]
    y_predicted_new=np.array(y_predicted_new)
    return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
log_loss(y_true,y_predicted)
```
