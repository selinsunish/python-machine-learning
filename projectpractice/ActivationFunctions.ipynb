```python
import math
def sigmoid(x):
    return 1/(1+math.exp(-x))
              
```

```python
#sigmoid function return always between 0 and 1
sigmoid(100)
```

```python
sigmoid(-56)
```

```python
def tanh(x):
    return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))
```

```python
tanh(-56)
```

```python
def leaky_relu(x):
    return max(0.1*x,x)
```

```python
leaky_relu(-100)
```
