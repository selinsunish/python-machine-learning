```python
#this projet is about guessing the prie give its size of the house
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
#used to split data into training and testing parts
from sklearn.model_selection import train_test_split
#Used to measure how wrong our predictions are.
from sklearn.metrics import mean_absolute_error
size=np.array([500,750,1000,1250,1500,1750,2000]).reshape(-1,1)
price=np.array([20,30,40,50,60,70,80])
df=pd.DataFrame({
    #.flatten() removes extra brackets.
    "Size(Sqft)":size.flatten(),
    "Price(Lakhs)":price
})
df
```

```python
#X_train	House sizes used for learning
#y_train	Prices used for learning
#X_test	New house sizes
#y_test	Real prices (hidden from model)
#20% data → testing

#80% data → training
x_train,x_test,y_train,y_test=train_test_split(size,price,test_size=0.3,random_state=42)
                                               
```

```python
model=LinearRegression()
#This is where learning happens.
#The computer:
#Looks at house sizes
#Looks at prices
#Finds a pattern
#Creates a math formula internally
# You never see the formula
# The computer creates it itself
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
error=mean_absolute_error(y_test,y_pred)
print("Mean Absolute Error:",error)
```

```python
#visualize results
plt.scatter(size,price)
plt.plot(size,model.predict(size))
plt.show()
plt.xlabel("House size (sqft)")
plt.ylabel("Price(Lakhs)")
plt.title("House Price Prediction")
plt.show()

```
